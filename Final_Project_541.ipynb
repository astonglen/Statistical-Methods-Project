<font size='5'>**3. METHODOLOGY**

<font size='4'>**Shapiro-Wilk Test**

<font size='4'>The Shapiro-Wilk test is a statistical test used to determine whether a dataset is normally distributed. The methodology involves calculating the test statistic, which is based on the covariance between the sample data and the normal probability plot. The p-value is then calculated using the null hypothesis that the sample data is normally distributed. If the p-value is less than the chosen significance level, the null hypothesis is rejected and it can be concluded that the sample data is not normally distributed.

<font size='4'>**Z-Test**

<font size='4'>The Z-test is a statistical test used to determine whether the mean of a sample is significantly different from a known population mean. The methodology involves calculating the test statistic Z, which is based on the difference between the sample mean and the population mean, normalized by the standard error of the mean. The p-value is then calculated using the null hypothesis that the sample mean is equal to the population mean. If the p-value is less than the chosen significance level, the null hypothesis is rejected and it can be concluded that the sample mean is significantly different from the population mean.

<font size='4'>**ANOVA**

<font size='4'>ANOVA (Analysis of Variance) is a statistical test used to determine whether there is a significant difference between the means of three or more groups. The methodology involves calculating the test statistic F, which is based on the ratio of the between-group variability to the within-group variability. The p-value is then calculated using the null hypothesis that the means of all groups are equal. If the p-value is less than the chosen significance level, the null hypothesis is rejected and it can be concluded that at least one group has a different mean than the others.

<font size='4'>**F-test**

<font size='4'>The F-test is a statistical test used to determine whether the variances of two populations are equal. The methodology involves calculating the test statistic F, which is based on the ratio of the variances of the two populations. The p-value is then calculated using the null hypothesis that the variances of the two populations are equal. If the p-value is less than the chosen significance level, the null hypothesis is rejected and it can be concluded that the variances of the two populations are not equal.

<font size='4'>**Tukey's Test**

<font size='4'>Tukey's test is a statistical test used to compare the means of multiple groups. The methodology involves calculating the test statistic Q, which is based on the difference between the means of each group, normalized by the standard error of the mean. The p-value is then calculated using the null hypothesis that the means of all groups are equal. If the p-value is less than the chosen significance level, the null hypothesis is rejected and it can be concluded that at least one group has a different mean than the others.

<font size='4'>**Chi-Squared Test for Independence**

<font size='4'>The chi-squared test for independence is a statistical test used to determine whether there is a significant association between two categorical variables. The methodology involves calculating the test statistic chi-squared, which is based on the difference between the observed frequencies and the expected frequencies, assuming that the two variables are independent. The p-value is then calculated using the null hypothesis that the two variables are independent. If the p-value is less than the chosen significance level, the null hypothesis is rejected and it can be concluded that there is a significant association between the two variables.

<font size='4'>**Linear Regression**

<font size='4'> Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. The methodology involves fitting a linear equation to the data using a least squares approach, where the coefficients of the equation represent the slope and intercept of the line. The goodness of fit of the model is evaluated using metrics such as the R-squared value.

<font size='4'>**Regularization**

<font size='4'>Regularization is a technique used in machine learning to prevent overfitting of a model. The methodology involves adding a penalty term to the objective function of the model, which penalizes large coefficients and encourages the model to be simpler. Common regularization techniques include L1 regularization (lasso) and L2 regularization (ridge regression).

<font size='4'>**K-Fold Cross Validation**

<font size='4'>K-fold cross validation is a technique used to evaluate the performance of a machine learning model by splitting the data into k equal subsets, or folds. In k iterations, k-1 folds will be used as training set and one will be used as validation set. K-fold cross validation can be used to assess the performance of a machine learning model and to compare the performance of different models or hyperparameters. It can help to prevent overfitting and provide a more accurate estimate of the model's performance on new data.

<font size='4'>**PCA**

<font size='4'>PCA (Principal Component Analysis) is a statistical technique used to reduce the dimensionality of a dataset while retaining as much of the original variation as possible. It involves finding the principal components, which are linear combinations of the original variables that explain the most variance in the data. PCA can be used for dimensionality reduction and visualization of high-dimensional data, as well as for feature extraction and data compression.

<font size='4'>**Polynomial Regression**

<font size='4'>Polynomial regression is a regression analysis technique that models the relationship between the independent variable x and the dependent variable y as an nth degree polynomial. Polynomial regression can be used to model non-linear relationships between variables, but it can also suffer from overfitting if the degree of the polynomial is too high.

<font size='4'>**Random Forest Regressor**

<font size='4'>Random forest is a machine learning algorithm that uses an ensemble of decision trees to make predictions. Random forest can be used for regression and classification tasks and is known for its high accuracy and ability to handle non-linear relationships and interactions between variables.

### ANOVA for total_stops

```python
import scipy.stats as stats

def calculate_f_statistic(*args):
    f_statistic, p_value = stats.f_oneway(*args)
    return f_statistic, p_value

alpha = 0.05
f_stat, p_val = calculate_f_statistic(zero_stop, one_stop, two_stop, three_stop)
print("F-test")
print('Statistics=%.3f, p=%.3f' % (f_stat, p_val))
if p_val < alpha:
    print("Reject Null Hypothesis, there is significant difference between all the samples.")
else:
    print("Fail to Reject Null Hypothesis, there is no significant difference between all the samples.")
```

### Tukeys Test

```python
import statsmodels.stats.multicomp as mc

samples = np.concatenate([zero_stop, one_stop, two_stop, three_stop])
group_names = ['zero_stop']*3475 + ['one_stop']*5625 + ['two_stop']*1317 + ['three_stop']*43
tukey_result = mc.pairwise_tukeyhsd(samples, group_names)
print(tukey_result)
```

### Forward Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model=LinearRegression()
def forward_selection():
    X = X_train
    y = y_train
    selected_features = []
    best_r2_score = 0
    for col in X.columns:
        selected_features.append(col)
        X_selected = X[selected_features]
        model = LinearRegression().fit(X_selected, y)
        y_pred = model.predict(X_selected)
        r2 = r2_score(y, y_pred)
        if r2 > best_r2_score:
            best_r2_score = r2
            print(best_r2_score)
        else:
            selected_features.remove(col)
    return selected_features

features = forward_selection()
features
```

### Backward Selection

```python
def backward_selection(X, y, metric):
    selected_features = list(X.columns)
    best_score = metric(y, LinearRegression().fit(X, y).predict(X))
    while len(selected_features) > 0:
        scores = []
        for feature in selected_features:
            features = [f for f in selected_features if f != feature]
            score = metric(y, LinearRegression().fit(X[features], y).predict(X[features]))
            scores.append((feature, score))
        best_feature, best_feature_score = max(scores, key=lambda x: x[1])
        if best_feature_score >= best_score:
            selected_features.remove(best_feature)
            best_score = best_feature_score
        else:
            break
    return selected_features

model = LinearRegression().fit(X_train, y_train)

selected_features = backward_selection(X_train, y_train, r2_score)
print(selected_features)

X_test = X_test[selected_features]
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R2 score on test set:",r2)
```

<font size=4>**4.5. RESAMPLING METHODS**

<font size=3>**K-FOLD CROSS VALIDATION**

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso, RidgeCV

num_folds = 5

alphas = [0.01, 0.1, 1, 10, 100]

kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
r2_scores = np.zeros((num_folds, len(alphas)))
for i, (train_idx, val_idx) in enumerate(kf.split(X_train, y_train)):
    X_train_fold, y_train_fold = X_train.iloc[train_idx], y_train.iloc[train_idx]
    X_val_fold, y_val_fold = X_train.iloc[val_idx], y_train.iloc[val_idx]
    for j, alpha in enumerate(alphas):
        model = Lasso(alpha=alpha).fit(X_train_fold, y_train_fold)
        y_val_pred = model.predict(X_val_fold)
        r2_scores[i, j] = r2_score(y_val_fold, y_val_pred)

mean_r2_scores = r2_scores.mean(axis=0)

best_alpha_idx = np.argmax(mean_r2_scores)
best_alpha = alphas[best_alpha_idx]

model = Lasso(alpha=best_alpha).fit(X_train, y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f'Lasso Regression - R2 score on test set: {r2:.2f}')

model = RidgeCV(alphas=best_alpha).fit(X_train, y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f'Ridge Regression - R2 score on test set: {r2:.2f}')
```
